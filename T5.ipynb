```python
from transformers import T5Tokenizer, T5ForConditionalGeneration
```

```python
model_name = 't5-small'
tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name)
```

```python
#para fazer resumo :"sumarize: texto a ser resumido"
text = "translate English to German: That is good."
input_ids = tokenizer.encode(text, return_tensors="pt", max_length=512, truncation=True)
```

```python
print(input_ids)
```

```python
tradutor_ids = model.generate(
    input_ids,
    max_length=40,
    #min_length=10,
    length_penalty=2.0,
    num_return_sequences=3,
    num_beams=4,
    early_stopping=True
    )
```

```python
print(tradutor_ids)
```

```python
tradutor_txt = tokenizer.decode(tradutor_ids[0], skip_special_tokens=True)
```

```python
print(tradutor_txt)
```
